```python
import pandas as pd
```

```python
dp = pd.read_csv("/content/LOTR.csv")
dp
```

```python
dp2 = pd.read_csv("/content/LOTR 2.csv")
dp2
```

```python
dp.merge(dp2,how='inner',on = ['FellowshipID'])  #inner join & merge
```

```python
dp.merge(dp2,how='outer')
```

```python
dp.merge(dp2,how='right',on = ['FellowshipID'])  #inner right
```

```python
dp.merge(dp2,how='left',on = ['FellowshipID'])  #inner left
```
